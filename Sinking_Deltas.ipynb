# Why are Delta's Sinking Faster than Sea Level is Rising? 
---

In 2009, Syvitski et al conducted a study, 'Sinking Delta's Due to Human Activities' that sought to explain this curious discrepancy between sinking deltas and rising sea levels. At the heart of this study is the acknowledgement that we, humans, have a big role to play in this phenomenon. With the installment of dams, reconstruction of distributary channels and processes like subsurface mining, we as humans have been influencing river and delta dynamics. Our impact has manifested itself into deltas that are sinking faster than sea level is rising due to a massive decrease or complete elimination of their aggradation rates. Sediment is being trapped by dams or not being properly transported through distributary channels becuase of our human infrastructure and as a result, the deltas are flooding and their vertical surface area above sea level is shrinking. In this data tutorial we will examine the various factors the Syvitski collected data for through different kinds of plots and seek to understand which factors influence the decreasing delta surface elevation.

# Relative Sea Level Rise
Lets continue looking at other factors to see how the relative sea level rise might compare across deltas. Create a plot that shows the min and max relative sea levels for all 33 deltas. Consider why we cannot simply plot the 'Relative Sea Level'.

Answer: We cannot use the 'Relative Sea Level' in its current state because it is a string and not an int or float. In the original data set the only value given was the relative sea level so I created the max and min ints in excel before loading the CSV into this data tutorial.

```python
import matplotlib.pyplot as plt

#plot each series
plt.plot(df['Min Relative sea-level rise (mm/yr)'])
plt.plot(df['Max Relative sea-level rise (mm/yr)'])

#display plot
plt.show()
```

This data isn't very clear - we need legends and titles to better understand what we are looking at. 

```python
#plot individual lines with custom colors and labels
plt.plot(df['Min Relative sea-level rise (mm/yr)'], label='Min', color='green')
plt.plot(df['Max Relative sea-level rise (mm/yr)'], label='Max', color='blue')

#add legend
plt.legend(title='Relative Sea Level Rise')

#add axes labels and a title
plt.ylabel('Relative Sea Level Rise', fontsize=14)
plt.xlabel('Delta', fontsize=14)
plt.title('Relative Min and Max Sea Level Rise for Each Delta', fontsize=16)

#display plot
plt.show()
```

This shows a large descrepancy between the min and max values...is there a benefit to using one over the other? Which value should be considered in delta recovery programs? For the sake of data visualization, it might be helpful to take an average of these two in order to get an average trend across the deltas. Let's now add a new column to our dataset that is the average of the min and max values for each delta. 

Answer: The min value might be more stable across the board as the graph shows and therefore might be a more accurate representation of the sea level rise. On the other hand, when considering recovery programs it is wise to use the worst case scenario so that the delta and surrounding landscape are prepaid for said worst case scenarios. While it might be an outlyer, the worst case scenario should be used when planning for disaster and delta recover.
